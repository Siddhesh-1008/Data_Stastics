# Measure of Central Tendency
It is a statistical measure used to find a single value that represents the center or typical value of a dataset. 
The main measures are Mean, Median and Mode.



Example 😄
Suppose class ke marks hain:
50, 55, 60, 65, 70

Teacher poochta hai:
Class ka overall performance kaisa hai?
Ab kya hum saare numbers bolenge?
50,55,60,65,70
❌ Nahi

Hume ek aisa number chahiye jo poore data ko represent kare.
Isi liye Measure of Central Tendency use karte hain.



# Measure of Central Tendency=Data ka center ya representative value
Ye batata hai:
Data generally kis value ke around hai?

Example 1 (Mean)
Marks:
50,55,60,65,70
Average:
(50+55+60+65+70)/5
= 300/5
= 60

So:
60 represents the entire class


Example 2 (Median)
Marks:
10,20,30,40,1000
Average:
220

😲 But class ke mostly marks 10-40 ke beech hain.

So Mean misleading ho gaya.
Middle value:

10,20,30,40,1000
      ↑
     30
Median = 30
Ye better representation ha

Example 3 (Mode)
Shirt sizes:
M,M,L,S,M,L
Most frequent:
M

Mode = M
Batata hai:
Most common value


#  MEASURE OF SPREAD 

Measure of Spread=Data kitna faila hua hai?(How much the data is spread out from center)
Values center se kitni door hain?

# Range (Data kitna spread hai(from smallest value to largest value))(OUTLIER SE AFFECT HOTA HAE)
Formula:
Range=Maximum - Minimum
Example
Marks:
10,20,30,40,50
Range:
50 - 10
= 40


# IQR(outlier se affect nhi hota)
Quartiles matlab data ko 4 equal parts me baantna.
10,20 | 30,40 | 50 | 60,70 | 80,90

# Q1 (25% point)
Lower half ka median:
10,20,30,40
Middle:
(20+30)/2 = 25
Q1 = 25

# Q3 (75% point)
Upper half:
60,70,80,90
Middle:
(70+80)/2 = 75
Q3 = 75

# Step 3: IQR

Formula:
IQR = Q3 - Q1
So:
75 - 25 = 50
IQR = 50

# IQR bolta hai:
Beech ke 50% students ke marks
25 se 75 ke beech me hain.
# IQR → sirf middle students dekhta hai (topper aur weakest ko ignore karke) ✅ BCZ IN RANGE WE TAKE HIGHEST -LOWEST IF THERE IS AN OUTLIER IN IT THEN IT WILL SHOW FALSE VAVLUES  JYADA SPREAD DIKHATA HAE PAR IN REAL IT IS NOT THAT MUCH

# OUTLIER
A value that is extremely different from the rest of the data.
Example:
40,45,50,55,500
Here:
500 is an Outlier.


# Effect on Mean 🚨
New Mean:
(40+45+50+55+500)/5
= 690/5
= 138

Mean changed from:
50 → 138
Huge change 😲

Why?
Because mean uses all values.
So outlier strongly affects Mean.

# Effect on Median 😄
Arrange data:
40,45,50,55,500
      ↑
Median:
50
Still same.

Why?
Median only looks at middle position.
It doesn't care if last value is:
60
or
500
or
5000
Middle remains 50.

# Effect on Standard Deviation 🚨

Without outlier:
40,45,50,55,60
Values close together.

Standard deviation:Small

With outlier:
40,45,50,55,500
Values very far apart.
Standard deviation:
Very Large

Why?
Std deviation measures:
How far values are from the mean
500 is extremely far.
So Std Dev increases a lot.

Measure	Effect of Outlier
Mean	Highly affected ❌
Median	Very little affected ✅
Mode	Usually not affected ✅
Standard Deviation	Highly affected ❌
Range	Highly affected ❌
IQR	Less affected ✅
