# What is Probability?
Probability measures the chance of occurrence of an event.
# Formula of Probability?
P(E)=Favorable Outcomes / Total Outcomes

# Independent Events?
One event does not affect another.
Example:
Coin Tosses


# Dependent Events?
One event affects another.

# Mutually Exclusive Events?
Cannot happen together.
Example:
Head and Tail in same toss




# Addition Rule
Used when question contains: OR

# Example:
# Red Card OR King
# General Case (Not Mutually Exclusive)

# Formula:
P(A ∪ B)=P(A)+P(B)-P(A ∩ B)

Why subtract?
Because common part counted twice.

Example 😄
Deck of cards:

Red Cards = 26
Kings = 4
Red Kings = 2

Formula:
26/52 + 4/52- 2/52=28/52


# Special Case (Mutually Exclusive)
If events cannot occur together:
P(A ∩ B)=0
Then formula becomes:

P(A ∪ B)=P(A)+P(B)
Example 😄

Die Roll

Event A:

Get 2

Event B:

Get 5
1/6 + 1/6

=

2/6
