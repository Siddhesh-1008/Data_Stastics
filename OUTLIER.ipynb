```python
import numpy as np 
import seaborn as sns
```

```python
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,3,201])
arr
```

# 5 NUMBER SUMMARY


```python
Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
print("Q1=>",Q1,"Q3=>",Q3)
```

```python
IQR=Q3-Q1
print("IQR=",IQR)
```

```python
# CALC UPPER FENCE AND LOWER FENCE 
UF=Q1+(1.5*IQR)
LF=Q3-(1.5*IQR)
print("UPPER FENCE:- ",UF ,"LOWER FENCE:- ",LF)
```

```python
# REMOVE OUTLIERS MEANS REMOVE NUMBERS THAT IS LEESS THAN UF AND BIGGER THAN UF
l=[]
for i in arr:
    if i<UF and i>LF:
        l.append(i)
arr2=np.array(l)
```

```python
# WITH OUTLIER IN BOX PLOT IF Y IS NOT THERE WE CAN JUST PLOT WITH THE HELP OF X VALUES
sns.boxplot(x=arr)
```

```python
# WITHOUT OUTLIER
sns.boxplot(x=arr2)
```
