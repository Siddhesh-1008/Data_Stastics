# What is Probability?
Probability measures the chance of occurrence of an event.
# Formula of Probability?
P(E)=Favorable Outcomes / Total Outcomes

# Independent Events?
One event does not affect another.
Example:
Coin Tosses

Example 😄
Coin tossed twice.

Event A:
1st toss = Head

Event B:
2nd toss = Tail

First toss ka result second toss ko affect nahi karta.

# Formula
P(A ∩ B)=P(A) × P(B)

Example:
1/2 × 1/2=1/4







# Dependent Events?
One event affects another.

Example 😄
52 cards deck.

Event A:
First card = Ace

Event B:
Second card = Ace

After first Ace:
Cards left = 51
Aces left = 3

Second probability changed.

# Dependent Events Formula P(A ∩ B)=P(A) × P(B|A)
Read as:
Probability of B given A






# Mutually Exclusive Events?
Cannot happen together.
Example:
Head and Tail in same toss
# Formula:- P(A ∩ B)=0





# Addition Rule
# Used when question contains: OR(∪)
# It is used to calculate the probability of two events occurring either or (OR condition).

# Example: Red Card OR King General Case (Not Mutually Exclusive)

# Formula:
P(A ∪ B)=P(A)+P(B)-P(A ∩ B)

Why subtract?
Because common part counted twice.

Example 😄
Deck of cards:

Red Cards = 26
Kings = 4
Red Kings = 2

Formula:
26/52 + 4/52- 2/52=28/52


# Special Case (Mutually Exclusive) AND(∩)
If events cannot occur together:
P(A ∩ B)=0
# Formula becomes:P(A ∪ B)=P(A)+P(B)
Example 😄

Die Roll

Event A:
Get 2

Event B:
Get 5

THEN 
1/6 + 1/6=2/6
