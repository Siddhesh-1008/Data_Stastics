# Hypothesis testing 
It is a statistical method used to determine whether there is enough evidence in a sample to support a claim about a population



# 2️⃣ Null Hypothesis (H₀):-Default assumption.
# Null hypothesis is the default assumption that there is no effect or no difference.

H₀ :
Coin Fair Hai

Matlab:
P(Head)=0.5
P(Tail)=0.5


# 3️⃣ Alternative Hypothesis (H₁) Opposite claim.
# Alternative hypothesis states that there is an effect, change or difference.

H₁ :
Coin Fair Nahi Hai
Matlab:
P(Head) ≠ 0.5


# Example
100 flips kiye.

Case 1:
Heads = 49
Tails = 51
Ye normal lag raha hai.
Coin fair hi lagta hai.

Case 2:
Heads = 95
Tails = 5
Ab suspicious hai 😄
Coin unfair lag raha hai.

# 5️⃣ Significance Level (α)
Sabse important.
Usually:α = 0.05

Meaning:-5% risk lene ko ready ho ki tum galat decision le sakte ho.

# 6️⃣ p-value
Suppose experiment ke baad mila:
p-value = 0.03

Rule:
If p-value < 0.05 Then Reject H₀
Meaning:
Coin fair nahi lag raha.

Agar:
p-value = 0.2 Then: Fail to Reject H₀

Meaning:
Coin fair ho sakta hai

# Confidence interval
Confidence Interval ka matlab:Ek estimate ki exact value nahi pata, isliye hum ek range batate hain.
Suppose sample se estimate mila:
If the parameter value specified in the null hypothesis lies within the confidence interval, we fail to reject the null hypothesis. If it falls outside the confidence interval, we reject the null hypothesis.

Head Probability = 0.52
95% Confidence Interval:
0.48 to 0.56

Meaning:
Hum 95% confident hain  ki actual head probability 0.48 aur 0.56 ke beech hai.

Confidence interval is calculated as:
Estimate ± Margin of Error.

# ONE TAILED TEST 
Only one direction check karni hai

College bolta hai:Average placement package = 10 LPA
Ye hamara Null Hypothesis (H₀) hai.
H₀ : Average package = 10 LPA

# Case 1: One-Tailed Test
Tum sirf ye check karna chahte ho:
Package BADHA hai kya?
Tumhe package kam hua ya same raha usse matlab nahi.

Then alternative hypothesis:H₁ : Average package > 10 LPA



# Two-Tailed Test
Dono directions check karni hain

Ab tum puchte ho:Package change hua hai kya?
Badh bhi sakta hai.
Kam bhi ho sakta hai.
Dono possibilities important hain.

Alternative:H₁ : Average package ≠ 10 LPA

Notice:
10 se zyada bhi
10 se kam bhi
Dono check karna hai.

Isliye:
Two-Tailed Test

# ZTEST USING P VALUES

```python
import numpy as np
from scipy.stats import norm
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
# POPULATION MEAN
population_mean=170
# sample mean 
sample_mean=np.mean(sample)
# STANDARD DEVIATION
std=3
n=len(sample)
# ZSCORE
z_score=(sample_mean-population_mean)/(std / np.sqrt(n))
z_score
```

```python
p_value=2*(1-norm.cdf(abs(z_score)))
p_value
```

```python
alpha=0.05
if(p_value < alpha):
    print("REJECT NULL HYPO AND ACCEPT ALTERANTE:- ")
else:
    print("ACCEPT NULL HYPO AND REJECT ALTERANTE:-")
```
