# ANNOVA TEST(ALSO CALLED AS F STATISTIC TEST)
# ANOVA (Analysis of Variance) is used to compare the means of three or more groups and 
# determine whether at least one group's mean is significantly different from the others.

```python
# IN THESE WE ARE COMPARING THE MEANS OF THREE GROUPS THAT IS FIRST,SECOND,THIRD CLASS 
# AND SEE WHETHER ALL CLASEES AVERAGE ARE SAME OR AT LEAST ONE GROUP AVERAGE DIFFERS FROM OTHER
```

import pandas as pd
import seaborn as sns
from scipy.stats import f_oneway

# Load Dataset
df = sns.load_dataset('titanic')

# Remove Null Values from age column
df = df.dropna(subset=['age'])

# Check Unique Passenger Classes
print(df['pclass'].unique())

# Separate Age Data for Each Class
class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']

# Perform One-Way ANOVA
f_stat, p_value = f_oneway(class_1, class_2, class_3)

print("F Statistic =", f_stat)
print("P Value =", p_value)

# Hypothesis Testing
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
    print("Average age differs across passenger classes")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant difference in average age")
