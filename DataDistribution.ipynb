# MEAN VARIANCE AND STD DEVIATION

# Why Use Mean?
Data ka center pata karne ke liye
Example:
Average salary
Average marks
Average age

# Why Use Variance?
Data kitna spread hai
Example:
Students consistent hain ya nahi

# Why Use Standard Deviation?
# Typical distance from mean yae basically bata hae pura data set ke values mean se kitna dur hae 
Example:

Mean Marks:
70

5

Means:
Most students roughly
65–75 ke around hain

```python
Mean=Center of data
Variance=Spread of data (squared)
Standard Deviation=Actual spread in original units
```

# FORMULAS

# Population
Mean μ = ΣX / N
Variance σ² = Σ(X - μ)² / N
Standard Deviation σ = √σ²


# Sample
Mean x̄ = ΣX / n
Variance s² = Σ(X - x̄)² / (n-1)
Standard Deviation s = √s²

# USE MEDIAN TO FIND CENTRAL TENDANCY AS OUTLER WONT AFFECT MEDIAN BUT IF WE USED MEAN THEN OUTLEIR CAN AFFECT IT
# Because skewed distributions often contain outliers that pull the mean. Median is resistant to outliers and gives a better measure of central tendency.

```python
Normal Distribution
→ Mean preferred ✅

Skewed Distribution
→ Median preferred ✅

Outliers present → Median

No Outliers → Mean
```
