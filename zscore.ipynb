# ZSCORE
# (SPECIFIC DATAPOINT OF DATASET HOW FAR IS IT FROM MEAN IN TERMS OD STD DEVIATION)
Ek specific value mean se kitni Std Deviations door hai
# EXAMPLE
Student marks:64
Z-score:(64 - 50)/7= 2
# Meaning:
Ye student mean se 2 standard deviations upar hai.

# Standard Deviation(ALL DATAPOINTS FROM THE DATASET HOW FAR IS IT FROM MEAN)

# Z-score tells:
1. How far a value is from mean

AND

2. What percentage of data lies below that value
(using Z-table)

# EXAMPLE OF ZCOREA ND Z SCORE TABLE
# FIND HOW FAR FROM MEAN
Mean (μ) = 8
Std Dev (σ) = 1
X(We need to find how far this value is it from mean) = 9.5

Formula:
Z = (X - μ)/σ
Z = (9.5 - 8)/1
= 1.5

Meaning:
Value 9.5 mean se 1.5 standard deviations upar hai if negative then niche hae woh 

# What percentage of data lies below or above that value
