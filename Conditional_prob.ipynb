# Bayes Theorem 
It is used to update the probability of an event when new evidence becomes available.

In simple words, it helps us find the probability of a cause after observing a result.
For example, after a person tests positive for Corona, Bayes theorem helps calculate the actual probability that the person truly has Corona.

# Normal Probability
Cause → Result

Example:
Rain
↓
Road Wet

# Bayes Theorem
Result → Cause

Example:
Road Wet
↓
Rain hone ka chance?
Bayes Theorem
Hum result dekhkar cause ka chance nikal rahe hain.
